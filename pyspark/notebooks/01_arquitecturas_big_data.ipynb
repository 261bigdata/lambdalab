## 1. ¿Qué es Big Data?

Big Data se refiere al procesamiento de grandes volúmenes de datos que no pueden ser manejados eficientemente con herramientas tradicionales. La sesión lo resume a partir de las **5V**: volumen, variedad, velocidad, veracidad y valor. 

## 2. Ecosistema Big Data

Una solución Big Data puede entenderse como un flujo de trabajo:

1. **Fuente de datos**
2. **Ingesta / extracción**
3. **Almacenamiento**
4. **Procesamiento**
5. **Visualización / consumo**

En la PPT se ilustra este ciclo con extracción, almacenamiento distribuido, procesamiento distribuido y visualización de valor. 

## 3. Batch vs Streaming

### Batch
Trabaja con datos acumulados y se procesa periódicamente.

**Ejemplos:** reportes diarios, logs históricos, entrenamiento de modelos.

### Streaming
Procesa datos conforme llegan al sistema.

**Ejemplos:** fraude, recomendaciones en tiempo real, monitoreo.

La sesión presenta ambos enfoques como base de decisión arquitectónica. 

## 4. Arquitectura Lambda

Lambda combina tres capas:

- **Batch layer**: procesamiento histórico
- **Speed layer**: procesamiento en tiempo real
- **Serving layer**: consulta de resultados

**Ventaja:** alta precisión al combinar histórico y tiempo real.

**Desventaja:** mayor complejidad.

## 5. Arquitectura Kappa

Kappa procesa todo como streaming.

**Ventaja:** arquitectura más simple.

**Desventaja:** menos optimizada para datos históricos.

## 6. Regla de decisión

- **Si necesitas histórico + tiempo real → Lambda**
- **Si todo es eventos en tiempo real → Kappa**

## 7. Comparación rápida

| Aspecto | Lambda | Kappa |
|---|---|---|
| Histórico | Sí | No separado |
| Tiempo real | Sí | Sí |
| Complejidad | Alta | Menor |
| Capas | Batch + Speed + Serving | Solo streaming |
| Reprocesamiento | Desde batch | Reprocesando stream |

Esta comparación aparece sintetizada en la presentación.
